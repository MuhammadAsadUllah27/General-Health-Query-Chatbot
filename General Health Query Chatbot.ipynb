```python
#here we import the google.generativeai
import google.generativeai as genai
```

```python
#we create an api key and here we give that key so that our model communicate with that api
genai.configure(api_key='YOUR_API_KEY')
```

```python
# Initialize the Gemini model
model = genai.GenerativeModel('gemini-2.5-flash')
```

```python
# Start a chat session and we use history to keep the record
chat = model.start_chat(history=[])

print("Hello! I am your Gemini-powered General Health Query chatbot. Type 'quit' to end the conversation.")
```

```python
#here we use loop so that the model works continuously untill user write "quit" in command
while True:
    user_input = input("You: ")
    if user_input.lower() == 'quit':
        print("Bot: Goodbye!")
        break

    # Here we tell the model that you are General Health Query chatbot and tell them to give the safety prompt
    prompt = f"You are a helpful General Health Query chatbot. Provide friendly and clear responses. Never give harmful medical advice and always recommend consulting a healthcare professional for diagnosis and treatment. \nUser: {user_input}\nMedical Assistant:"
    try:
        response = chat.send_message(
            prompt,
            generation_config={'temperature': 0.9}
        )
        bot_response = response.text
    except Exception as e:
        bot_response = f"An error occurred: {e}"

    print(f"Bot: {bot_response}")
```
